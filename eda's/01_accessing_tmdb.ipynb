**//IMPORTS**

```python
import requests
import pandas as pd
import datetime as dt
import os, sys, csv, calendar, time
```

**//CONFIGS**

```python
API_KEY = os.getenv("TMDB_API_KEY")
print(repr(API_KEY))

URL = "https://api.themoviedb.org/3/discover/movie"
URL_DISC   = "https://api.themoviedb.org/3/discover/movie"
URL_DETAIL = "https://api.themoviedb.org/3/movie/{}" 

headers = {"Authorization": f"Bearer {API_KEY}"}

YEARS = [2024]
```

**//MOVIE LOOP**

```python
movies = []
page, pages, rows = 1, 1, []
for year in YEARS:
    for month in range(1, 2):
        start = dt.date(year, month, 1)
        last  = calendar.monthrange(year, month)[1]
        end   = min(dt.date(year, month, last), dt.date.today())
        if start > dt.date.today():
            break                                   # don’t query future months

        page, pages = 1, 1
        while page <= pages:
            params = {
                "api_key": API_KEY,
                "primary_release_date.gte": start,
                "primary_release_date.lte": end,
                "sort_by": "primary_release_date.asc",
                "page": page,
                "include_adult": "false"
            }
            data = requests.get(URL_DISC, params=params, timeout=20).json()
            pages = data.get("total_pages", 1)

            for m in data["results"]:
                movie_id = m["id"]
                # second request just for the IMDb ID
                d = requests.get(URL_DETAIL.format(movie_id),
                                 params={"api_key": API_KEY},
                                 timeout=10).json()
                rows.append([
                    m["title"],
                    m["release_date"],
                    d.get("imdb_id")           # might be None for very new films
                ])

            print(f"{year}-{month:02d} p{page}/{pages} ✅")
            page += 1
            time.sleep(0.2)                    # courtesy delay (~5 req/s)

outfile = "tmdb_movies_2024_with_imdb.csv"
with open(outfile, "w", newline="", encoding="utf-8") as f:
    csv.writer(f).writerows([["title", "release_date", "imdb_id"]] + rows)

print(f"Saved {len(rows)} rows to {outfile}")
```
